Using matplotlib plot the following functions on the domain: x = 0.5 to 100.0 in steps of 0.5.
1) y=x
2) y=$x^2$
3) y=$(x^3)/100$
4) y=$sin(x)$
5) y=$(sin(x))/x$
6) y=$logx$
7) y=$e^x$

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#for y=x
x = np.arange(0.5,100.5,0.5)
y1 = x
plt.plot(x,y1, label = "y=x")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Plot of y=x")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#for y=x^2
def function_2(x):
    return x**2
x = np.arange(0.5,100.5,0.5)
y2 = function_2(x)
plt.plot(x,y2, label = "y=x^2")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Plot of y=x^2")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#for y=(x^3)/100
def function_3(x):
    return (x**3)/100
x = np.arange(0.5,100.5,0.5)
y3 = function_3(x)
plt.plot(x,y3, label = "y=(x^3)/100")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Plot of y=(x^3)/100")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#for y=sin(x)
def function_4(x):
    return np.sin(x)
x = np.arange(0.5,100.5,0.5)
y4 = function_4(x)
plt.plot(x,y4, label = "y=sin(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Plot of y=sin(x)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#for y=sin(x)/x
def function_5(x):
    return (np.sin(x))/x
x = np.arange(0.5,100.5,0.5)
y5 = function_5(x)
plt.plot(x,y5, label = "y=sin(x)/x")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Plot of y=sin(x)/x")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#for y=log(x)
def function_6(x):
    return np.log(x)
x = np.arange(0.5,100.5,0.5)
y6 = function_6(x)
plt.plot(x,y6, label = "y=log(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Plot of y=log(x)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#for y=e^x
def function_7(x):
    return np.exp(x)
x = np.arange(0.5,100.5,0.5)
y7 = function_7(x)
plt.plot(x,y7, label = "y=e^x")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Plot of y=e^x")
plt.legend()
plt.grid(True)
plt.show()
```
